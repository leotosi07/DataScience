**Hotel Reservations Dataset**

Los canales de reserva de hoteles en línea han cambiado drásticamente las posibilidades de reserva y el comportamiento de los clientes, la base tiene una recopilacion de datos de estos clientes que realizan las reservas.

Glosario:

0. Booking_ID: identificador único de cada reserva
1. no_of_adults: Número de adultos
2. no_of_children: Número de niños
3. no_of_weekend_nights: Número de noches de fin de semana (sábado o domingo) que el huésped se hospedó o reservó para quedarse en el hotel
4. no_of_week_nights: Número de noches de la semana (lunes a viernes) que el huésped se hospedó o reservó para quedarse en el hotel
5. type_of_meal_plan: Tipo de plan de comidas reservado por el cliente:
6. required_car_parking_space: ¿Necesita el cliente una plaza de aparcamiento? (0 - No, 1- Sí)
7. room_type_reserved: Tipo de habitación reservada por el cliente. Los valores son cifrados (codificados) por INN Hotels.
8. lead_time: Número de días entre la fecha de reserva y la fecha de llegada
9. arrival_year: año de la fecha de llegada
10. arrival_month: Mes de la fecha de llegada
11. arrival_date: Fecha del mes
12. market_segment_type: Designación del segmento de mercado.
13. repeated_guest: ¿El cliente es un invitado repetido? (0 - No, 1- Sí)
14. no_of_previous_cancellations: número de reservas anteriores que fueron canceladas por el cliente antes de la reserva actual
15. no_of_previous_bookings_not_canceled: Número de reservas anteriores no canceladas por el cliente antes de la reserva actual
16. avg_price_per_room: Precio medio por día de la reserva; Los precios de las habitaciones son dinámicos. (en euros)
17. no_of_special_requests: número total de solicitudes especiales realizadas por el cliente (por ejemplo, piso alto, vista desde la habitación, etc.)
18. booking_status: Flag que indica si la reserva fue cancelada o no.

La pregunta principal es, con estos datos, se puede llegar a obtener un indice de cancelacion o compromiso del cliente, se puede determinar que clientes son los que mayormente cancelan, para orientar el negocio a los otros tipos de clientes.
